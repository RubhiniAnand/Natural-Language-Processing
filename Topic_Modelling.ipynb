```python
# import the necessary libraries
import pandas as pd
```

```python
#load the document
df = pd.read_csv("GPTReviews.csv",header=None,on_bad_lines='skip',encoding='utf8')
```

```python
df
```

```python
#retrieve the 1st 100 rows
df=df.iloc[0:100]
```

```python
# drpping the unwanted columns
cols_to_del=[1,4,5,6,7]
df=df.drop(cols_to_del,axis=1)
df=df.drop(0,axis=0)
```

```python
#renaming the columns
df.columns=["Name","Comment","Date"]
```

```python
date=df['Date'].str.split(' ',n=1,expand=True)
df["Date"]=date[0]
df.dropna(inplace=True)
```

```python
df
```

```python
#extracting features using Hashing Vectorizer
from sklearn.feature_extraction.text import HashingVectorizer
vectorizer = HashingVectorizer(n_features=1000, stop_words='english',alternate_sign=False)
hashed_features = vectorizer.fit_transform(df['Comment'])
```

```python
print(hashed_features)
```

## 456 stored elements: The total number of non-zero elements in the matrix.
## Shape (99, 500): The matrix has 99 rows and 500 columns.
## The pairs (row, column) are the coordinates of the non-zero values in the matrix.
## (0, 187) 0.4472135954999579: At row 0 and column 187, the value is approximately 0.4472.
## (1, 242) 1.0: At row 1 and column 242, the value is 1.0.

```python
# LDA Model
from sklearn.decomposition import LatentDirichletAllocation
lda = LatentDirichletAllocation(n_components=5, random_state=0)
lda_model=lda.fit_transform(hashed_features)
```

## Each row in the output represents a document, and each column represents a topic.
## First Row [0.06434561, 0.06212836, 0.74679681, 0.06231034, 0.06441888]:
This represents the topic distribution for the first document.
The document has approximately a 74.68% probability of belonging to the third topic, with much smaller probabilities for the other topics.

```python
# The components_ matrix gives insight into which words are most associated with each topic.
lda.components_
```

```python
#the topic with highest probability is returned
df.loc[:,"Topic"]=lda_model.argmax(axis=1)
```

```python
df
```

```python
df["Topic"].value_counts()
```

# COUNT VECTORIZER

```python
#renaming the dataset
dataset=df
```

```python
dataset
```

```python
#extracting features using Count Vectorizer
from sklearn.feature_extraction.text import CountVectorizer
count_vector=CountVectorizer(stop_words='english')
lda_count=count_vector.fit_transform(dataset['Comment'])
print(lda_count)
```

```python
#feature names from the document
count_vector.get_feature_names_out()
```

```python
#total no. of feature names extracted
len(count_vector.get_feature_names_out())
```

```python
# array format of the document after counter vector
print(lda_count.toarray())
```

```python
#Applying the LDA model
from sklearn.decomposition import LatentDirichletAllocation
lda2 = LatentDirichletAllocation(n_components=5, random_state=0)
lda_count_model=lda2.fit_transform(lda_count)
```

```python
#printing the clustered words inside the 5 topics
for index,Topic in enumerate(lda2.components_):
    results=([count_vector.get_feature_names_out()[i] for i in Topic.argsort()[-10:]])
    print(results)
```

```python
Topic.shape
```

```python
#taking out the most likely topic for each document and assign it to the dataframe
dataset.loc[:,"Topic1"]=lda_count_model.argmax(axis=1)
```
