```python
# import the necessary libraries
import pandas as pd
```

```python
# load the document
df = pd.read_csv("sampleChat.txt",header=None,on_bad_lines='skip',encoding='utf8')
```

```python
df
```

```python
#Dropping and renaming the column
df=df.drop(0)
columntoDel=[2,3]
df=df.drop(columns=columntoDel,axis=1)
df.columns=['Date','Chat']
```

```python
#preprocessing the date time column
df['Date'] = pd.to_datetime(df['Date'], format='mixed',errors='coerce')
start_date = '2024-04-18 00:00:00'
end_date = '2024-05-25 23:59:59'
df = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]
```

```python
df
```

```python
#splitting the time,name and chat sa a individual columns
msg=df['Chat'].str.split('-',n=1,expand=True)
msg1=msg[1].str.split(':',n=1,expand=True)
df.loc[:,'Time']=msg[0]
df.loc[:,'Name']=msg1[0]
df.loc[:,'Chat']=msg1[1]
df=df[['Date','Time','Name','Chat']]
df
```

```python
# Dropping the unwanted texts
df = df.copy()
df.drop(df[df['Chat'].str.contains(r'<\s*Media\s+omitted\s*>', na=False)].index, inplace=True)
```

```python
#defining a function to perform Sentimental Analysis
def sentimentalAnalysis(data,columnname):
        
        #downloading vader_lexicon for the process
        import matplotlib.pyplot as plt
        import seaborn as sns
        "Importing Necessary Packeage"
        from nltk.sentiment.vader import SentimentIntensityAnalyzer
        sid = SentimentIntensityAnalyzer()
        
        "Deleting null pr empty value"
        data.dropna(inplace=True)
        
        "Checking for a comment"
        #sid.polarity_scores(data[columnname][93])
        
        "Creating respective columns"
        data.loc[:, 'scores'] = data[columnname].apply(lambda commentText: sid.polarity_scores(commentText))
        data.loc[:, 'compound']  = data['scores'].apply(lambda score_dict: score_dict['compound'])
        data.loc[:, 'Positive'] = data['scores'].apply(lambda score_dict: score_dict['pos'])
        data.loc[:, 'Negative']=data['scores'].apply(lambda score_dict: score_dict['neg'])
        data.loc[:, 'Neutral']  = data['scores'].apply(lambda score_dict: score_dict['neu'])
        
        "Creating final pos or neg using compound score"
        data.loc[:, 'comp_score'] = data['compound'].apply(lambda c: 'pos' if c >=0 else 'neg')
        plt.clf()
        #comp=sns.countplot(x = 'comp_score', hue = 'Name', data = data, palette = 'magma')
        #comp.figure.savefig("date_charts.png")
        "Checking how many pos and neg"
        posneg=pd.DataFrame(data['comp_score'].value_counts())
        return posneg,data
    
```

```python
#calling the defined function
pos,data_Senti=sentimentalAnalysis(dataset,columnname='Chat')
```

# Topic Modelling

```python
#importing TFIDF Vectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
#perform Tfidf vectorizer 
tfidf=TfidfVectorizer(max_df=0.95,min_df=2,stop_words='english')
dtm=tfidf.fit_transform(df["Chat"])
```

```python
tfidf
```

```python
#Model is created on the preprocessed chat column
from sklearn.decomposition import NMF
nmf_model=NMF(n_components=5,random_state=42)
nmf_model.fit(dtm)
```

```python
nmf_model.components_
```

```python
# higher weightage words after ifidf vectorization
tfidf.get_feature_names_out()
```

```python
# #printing the clustered words inside the 5 topics
for index,topic in enumerate(nmf_model.components_):
    results=([tfidf.get_feature_names_out()[i] for i in topic.argsort()[-10:]])
    print(results)
```

```python
# tranforming the model created column
topic_results=nmf_model.transform(dtm)
```

```python
topic_results
```

```python
#taking out the most likely topic for each document and assign it to the dataframe
df.loc[:,"Topic"]=topic_results.argmax(axis=1)
```
